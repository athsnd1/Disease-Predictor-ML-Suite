```
import pandas as pd
df = pd.read_excel('/content/drive/MyDrive/Original Symptoms_data.xlsx')
```

```
#checking for missing values:
print(df.isnull().sum())

#if result is 0 for all columns, there are no empty rows
```

```
#1. creating the sum of the chosen symptoms:
key_features = ['CTRH', 'FTG', 'FVR', 'HDACH', 'HGGDFVR', 'NUS', 'SRTRT']
df['sum'] = df[key_features].sum(axis=1)
#df.head()

#2. defining a threshold using the median:
threshold = df['sum'].median()

#3. creating the binary target column based on threshold
df['target'] = (df['sum']>threshold).astype(int)

#4. dropping the sum so the model doesn't cheat
x = df[key_features]
if 'total_load' in key_features:
    x = x.drop('total_load', axis=1)
y = df['target']

# Creating a copy of the features
x_binary = x.copy()

# Binarize: Any value >= 3 becomes 1 (High), others 0 (Low/None)
x_binary = (x_binary >= 3).astype(int)

print("Features shape:", x.shape)
print("Target shape:", y.shape)
```

```
#checking the distribution of the target column:
print(df['target'].value_counts())
```

```
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Splitting the data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x_binary, y, test_size=0.2, random_state=42)

print("Training set shape:", x_train.shape)
print("Test set shape:", x_test.shape)

# Initializing the Logistic Regression model
model = LogisticRegression()

# Training the model
model.fit(x_train, y_train)

# Making predictions on the test set
y_pred = model.predict(x_test)

# Printing the classification report
print(classification_report(y_test, y_pred))
```

```
# Create a DataFrame of the coefficients for the binary model
binary_coeffs = pd.DataFrame({'Feature': key_features, 'Coefficient': model.coef_[0]})
print(binary_coeffs.sort_values(by='Coefficient', ascending=False))
```
