```
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```
df = pd.read_excel('/content/drive/MyDrive/Original Symptoms_data.xlsx')
```

```
df.info()
```

```
df.describe()
```

```
df.columns
```

```
print(df.isnull().sum()) #no null values
```

```
key_features = ['FVR','NUS','FTG','VMT','CHLNRIG','HDACH']
df['sum'] = df[key_features].sum(axis=1)

threshold = df['sum'].median()

df['target'] = (df['sum']>threshold).astype(int)
x = df[key_features]
y = df['target']

print("Feature shape:", x.shape)
print("Target shape:", y.shape)
print("Class balance of target label:",y.value_counts())
```

```
x_binary = x.copy()

x_binary = (x_binary>=3).astype(int)

#splitting the model into training and test data
x_train, x_test, y_train, y_test = train_test_split(x_binary,y,test_size=0.2,random_state=42)
print("Training set shape:", x_train.shape)
print("Test set shape:", x_test.shape)
```

```
#initializing the model:
model = LogisticRegression()

#traing the model:
model.fit(x_train,y_train)

#predict:
y_pred = model.predict(x_test)

#evaluate
print(classification_report(y_test,y_pred))
```
